   The assignment will use data generated in the API Challenge assignment and display it on a webpage, but the data needs a
   little massaging, first.
   
   This notebook reads in the weather cities CSV file, converts the index to a column so it behaves like the rest of the
   columns on our website, rearranges the column order so that it makes sense to the web user, and writes the dataframe
   in HTML format.

```python
# Dependencies and Setup
import pandas as pd


# set input and output filenames
input_data_file = "Resources/cities.csv"
output_data_file = "Resources/cities.html"
```

```python
# read CSV file into dataframe and display
df = pd.read_csv(input_data_file)
df
```

```python
# convert the index to a column in the dataframe
# and display
df["City ID"] = df.index
df
```

```python
# rearrange the column order for display on our web page
# and display the new dataframe
df = df[["City ID", "City", "Country", "Lat", "Lng", "Date", "Max Temp", "Humidity", "Cloudiness", "Wind Speed"]]
df
```

```python
# write the file to HTML format
df.to_html(output_data_file)
```
